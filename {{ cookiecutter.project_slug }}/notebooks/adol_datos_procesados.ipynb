Una vez llamadas correctamente las columnas y seleccionados los datos, podemos concatenar los dataframes y obtener uno solo.

```python
# Lista de DataFrames
dataframes = [filtered_df_2006, filtered_df_2012, filtered_df_2018, filtered_df_2020, filtered_df_2021, filtered_df_2022,filtered_df_2023]  # Reemplaza con los nombres de tus DataFrames

# Concatenar los DataFrames hacia abajo
df_concatenado = pd.concat(dataframes, axis=0, ignore_index=True)
```

Veamos cuantos datos perdidos tenemos:

```python
df_concatenado.isna().sum()
```

Como podemos ver hay muchos datos faltantes en muchos datos en las ultimas columnas, esto se debe a que las encuestas anteirores al 2021 no trataban la depresion ni el sentimiento de tristeza como parte del cuestionario.


Sustituimos los datos faltantes con $0$, y convertimos nuestras columnas a un tipo de dato correcto.

```python
df_concatenado[['Depresion','Tristeza']] = df_concatenado[['Depresion','Tristeza']].fillna(0)


# Convertir una columna específica a categoría
df_concatenado['Sexo'] = df_concatenado['Sexo'].astype('category')
df_concatenado['Entidad'] = df_concatenado['Entidad'].astype('category')
df_concatenado['Atentar_contras_si'] = df_concatenado['Atentar_contras_si'].astype('category')
df_concatenado['Depresion'] = df_concatenado['Depresion'].astype('int').astype('category')
df_concatenado['Tristeza'] = df_concatenado['Tristeza'].astype('int').astype('category')
df_concatenado['Fecha'] = pd.to_datetime(df_concatenado['Fecha'])
```
