# Analizar los catalogos


Siguiendo un analisis similar al de los catalogos de adolescentes contruimos los siguientes dataframes

# 2006

```python
import pandas as pd
import os

# Ruta del archivo
file_path = os.path.join('..', 'data', 'raw', 'DATOS ADULTOS', 'ENSANUT-Adultos-Datos-2006.csv')

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(file_path, low_memory=False)
df['Fecha'] = pd.to_datetime('2006-07-01')
# Seleccionar las columnas que te interesan
filtered_df_2006 = df[['edad', 'sexo', 'ent','Fecha', 'a301a']]
# Renombrar columnas específicas
filtered_df_2006 = filtered_df_2006.rename(columns={'edad': 'Edad', 'sexo': 'Sexo', 'ent':'Entidad', 'a301a':'Tristeza'})

# Imprimir las filas filtradas
filtered_df_2006
```

# 2023

```python
file_path = os.path.join('..', 'data', 'raw','DATOS ADULTOS', 'ENSANUT-Adultos-Datos-2023.csv')

# Cargar el archivo Excel en un DataFrame, omitiendo las primeras 16 filas
df = pd.read_csv(file_path, sep=';')
# Seleccionar las columnas que te interesan
filtered_df_2023 = df[['edad', 'sexo', 'entidad','fecha_ini','a1213', 'a0213','a0217']]
filtered_df_2023 = filtered_df_2021.rename(columns={'edad': 'Edad', 'sexo': 'Sexo', 'entidad': 'Entidad', 'fecha_ini': 'Fecha', 'a1213': 'Atentar_contras_si', 'a0213': 'Depresion', 'a0217': 'Tristeza'})

# Imprimir las filas filtradas
filtered_df_2023
```

Ahora los podemos concatenar

```python
# Lista de DataFrames
dataframes = [filtered_df_2006, filtered_df_2012, filtered_df_2018, filtered_df_2020, filtered_df_2021, filtered_df_2022,filtered_df_2023]  # Reemplaza con los nombres de tus DataFrames

# Concatenar los DataFrames hacia abajo
df_concatenado = pd.concat(dataframes, axis=0, ignore_index=True)
```

Vemos los datos faltantes

```python
df_concatenado.isna().sum()
```

Sustituimos los datos faltantes con $0$, y convertimos nuestras columnas a un tipo de dato correcto.

```python
df_concatenado[['Atentar_contras_si','Depresion','Tristeza']] = df_concatenado[['Atentar_contras_si','Depresion','Tristeza']].fillna(0)


#Convertir una columna específica a categoría
df_concatenado['Sexo'] = df_concatenado['Sexo'].astype('category')
df_concatenado['Entidad'] = df_concatenado['Entidad'].astype('category')
df_concatenado['Atentar_contras_si'] = df_concatenado['Atentar_contras_si'].astype('category')
#df_concatenado['Depresion'] = df_concatenado['Depresion'].astype('int').astype('category')
#df_concatenado['Tristeza'] = df_concatenado['Tristeza'].astype('int').astype('category')
df_concatenado['Fecha'] = pd.to_datetime(df_concatenado['Fecha'])
```

Verificamos los tipos

```python
# Verificar tipos de datos de todas las columnas
print(df_concatenado.dtypes)
```
